# Usabe of Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 9, 16, 25]
y2 = [1, 4, 9, 16, 25]
```

## 1. Line Plot

```python
plt.figure(1)  # create first figure
plt.plot(x, y, label="y = 2x", color= "blue", marker = "o")
plt.plot(x, y2, label = "y = x^2", color = "red", linestyle = "--")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot Example")
plt.legend()
```

This means that:
- Each (x, y) data point will be displayed as a blue circle
- These circles will be connected by a solid blue line (the default line style)
- The circles make it easier to identify the exact position of each data point

## More about markers in Matplotlib

Matplotlib offers many different marker styles:
- `'o'` - circle
- `'.'` - point
- `'s'` - square
- `'^'` - triangle up
- `'v'` - triangle down
- `'*'` - star
- `'+'` - plus
- `'x'` - x

You can also control marker properties with additional parameters:
- `markersize` or `ms` - controls the size of markers
- `markerfacecolor` or `mfc` - controls the fill color of markers
- `markeredgecolor` or `mec` - controls the edge color of markers
- `markeredgewidth` or `mew` - controls the edge width of markers

For data visualization, markers are particularly useful when:
- You have a small to moderate number of data points
- You want to emphasize individual data points
- You need to distinguish between multiple data series

Useful Python packages for plotting with markers include matplotlib, seaborn, and plotly.

## 2. Bar Chart

```python
plt.figure(2)  # create second figure
students = ["Aparna", "Monica", "Rohit", "Sagar"]
marks = [85, 70, 90, 60]

plt.bar(students, marks, color="green")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Bar Chart Example")
```

### 3. Scatter Plot

```python
plt.figure(3)  # create third figure
plt.scatter(x, y2, color="purple")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Scatter Plot Example")
```

### Show all plots

```python
plt.show()
```
